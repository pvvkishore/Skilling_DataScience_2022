```python
token = Tokenizer()
token.fit_on_texts([data])
encoded = token.texts_to_sequences([data])[0]
#tokenizer.fit_on_texts(data2)
#encode1 = token.texts_to_matrix(data1, mode='count')
print(len(encoded))
print(data)
```

```python
vocab_size = len(token.word_index) + 1
print('Vocabulary Size: %d' % vocab_size)
```

```python
sequences = list()
```

```python
for i in range(1, len(encoded)):
	sequence = encoded[i-1:i+15]
	sequences.append(sequence)
print('Total Sequences: %d' % len(sequences))
print(sequences)
```

```python
sequences=tf.keras.preprocessing.sequence.pad_sequences(sequences,maxlen=15, padding="post")
#print(sequences)
```

```python
sequences = np.array(sequences)
print(sequences[:])
X, y = sequences[:,0],sequences[:,1]
print(y)
```

```python
#X
```

```python
#y
```

```python
y = to_categorical(y, num_classes=vocab_size)
```

```python
# define model
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=1))
```

```python
model.add(LSTM(50, activation='relu'))
model.add(Dense(vocab_size, activation='softmax'))
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit network
model.fit(X, y, epochs=500, verbose=1)
```

# Text Prediction

```python
n_words = 15
text_test = 'Jack'
result = 'Jack'
```

```python
for _ in range(n_words):
    encoded = token.texts_to_sequences([text_test])[0]
    encoded = np.array(encoded)
    # predict a word in the vocabulary
    y_Predict = model.predict_classes(encoded, verbose=0)
    Next_word = ''
    for word, index in token.word_index.items():
        if index == y_Predict:
            Next_word = word
            break
    text_test, result = Next_word, result + ' ' + Next_word
```
