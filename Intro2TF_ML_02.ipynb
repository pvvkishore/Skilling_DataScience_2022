```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(units=1, input_shape=[1])])
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
```

```python
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

```python
model.fit(xs,ys, epochs=500)
```

# How Computers Learn to Find That Relationship.
# At the start The Computer has no Idea about the realtionship between X and Y.
# So It Guess Randomly. Lets Assume: Y = 10X+10 
# To Measure the Goodness or Badness in the guess work done by the computer you need a target Y(The Actual Y)
# THe Guess is the Predicted Y.
# The LOSS Function Decides the G or B in the Guess Work

# Substitute X values in Y = 10X+10
# For Ex. X= -1, Yp = 0, Ya = -3 
# Mean Square Error = sum((Yp - Ya)**2) for All X values
# The JOB OF LOSS FUNCTION

# What is the Next Guess Y = 20X+10 or Y = 5X+5 or Y = 2X-5 or ....
# How to decide which is next best guess 
# The JOB of the OPTIMIZER

### Print Weights

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

lo = Dense(units=1, input_shape = [1])
model = Sequential(lo)
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
model.fit(xs,ys, epochs = 500)

print(model.predict([10.0]))
print("This is what I have learned: {}".format(lo.get_weights()))
```
